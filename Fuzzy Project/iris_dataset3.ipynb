```python
iris = datasets.load_iris()
iris.target[149]

iris.target_names

type(iris)

data = pd.DataFrame(iris.data)
cols = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

data.head()

min('petal_length')

data['species'] = iris.target

data.head(150)

data.loc[::50]

data.describe()

data.species.value_counts()

for k in range(len(cols)):
    print(f"feature {k}: {cols[k]}")
    print(f"\tclass 0 (blue) min: {np.min(data[k][:50])}, class 0 max: {np.max(data[k][:50])}")
    print(f"\tclass 1 (orange) min: {np.min(data[k][50:100])}, class 1 max: {np.max(data[k][50:100])}")
    print(f"\tclass 2 (green) min: {np.min(data[k][100:150])}, class 2 max: {np.max(data[k][100:150])}\n")

fig, ax = plt.subplots(2,2, figsize = (9, 5))

class_labels = [0] * 50 + [1] * 50 + [2] * 50

k = 0
for i in range(2):
    for j in range(2):
        ax[i, j].set_xlabel("value (length/width in cm)")
        ax[i, j].set_ylabel("distribution")
        #ax[i, j].set_ylim(0,45)
        ax[i, j].hist([data[k][:50], data[k][50:100], data[k][100:150]], bins = 15, label = [class_labels[49], class_labels[99], class_labels[149]])
        ax[i, j].legend()
        ax[i, j].set_title(f"{cols[k]}",fontsize=14)
        k = k + 1
plt.tight_layout()
```
