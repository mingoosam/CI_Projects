```python
classes = np.linspace(0, 100, 50)
# fig,ax = plt.subplots(2,2)

abcd_C0 = [0, 10, 20, 33]
membership_C0 = fuzz.trapmf(classes, abcd_C0)

abcd_C1 = [30, 40, 50, 67]
membership_C1 = fuzz.trapmf(classes, abcd_C1)

abcd_C2 = [65, 75, 85, 100]
membership_C2 = fuzz.trapmf(classes, abcd_C2)

fig, ax = plt.subplots(figsize=(8,5))
ax.plot(classes,membership_C0,"orange",label="Low Confidence")
ax.plot(classes,membership_C1,"purple",label="Medium Confidence")
ax.plot(classes,membership_C2,"green",label="High Confidence")
ax.set_xlabel("Confidence")
ax.set_ylabel("Class 0")
ax.set_title("Consequence Membership Functions")
ax.legend(bbox_to_anchor=(1,1),loc="lower center")
#plt.margins(0.8)
fig.tight_layout()
fig.savefig('memberships.png')
```
