```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fitness(x, y):
    
    return (x ** 2 + y - 11) ** 2 + (x + y ** 2 - 7) ** 2

    #return 0.26 * (x ** 2 + y ** 2) - 0.48 * (x * y)
    #return x * 2 + 4 * y ** 2
```

```python
# Create worlds (2D, 3D)

x_vals = np.arange(-5, 5, 0.1)
y_vals = np.arange(-5, 5, 0.1)

xx, yy = np.meshgrid(x_vals, y_vals)

zz = fitness(xx, yy)

# Example scatter plots that can be shown in world

x_s = np.asarray([np.random.randint(-4, 4) for i in range(5)])
y_s = np.asarray([np.random.randint(-4, 4) for i in range(5)])
z_s = fitness(x_s, y_s)
```

```python
fig, ax = plt.subplots(figsize = (7, 6))
ax.contourf(xx, yy, zz, levels=np.logspace(-9, 9, 50), cmap='jet', alpha=0.4)
ax.scatter(x_s, y_s, fitness(x_s, y_s), c = "black")
```
