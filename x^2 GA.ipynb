```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
# Create chromosome blueprint

class Chromosome:
    
    def __init__(self, x):
        self.x = x
        self.fitness = None
```

```python
# Create population of agents

def create_population(num_chromosomes):
    
    population = []
    for i in range(num_chromosomes):
        population.append(Chromosome(np.random.randint(-1000, 1000)))
        #print(population[i].x)
    return population 

#pop = create_population(10)
```

```python
# Create way to display chromosomes 

def display_population(population, string):
    
    print(f"\nPopulation after {string}:\n")
    for i, chromosome in enumerate(population):
        print("Chromosome %s : x = %s, fitness = %s" % (i, chromosome.x, chromosome.fitness))
```

```python
# Create fitness function, f(x) = x^2
# maps a chromosome representation into a scalar value

def fitness(x):
    
    return x ** 2

# Create function to update each chromosome's fitness

def evaluate_population(population):

    for chromosome in population:
        chromosome.fitness = fitness(chromosome.x)
        
    scores = [chromosome.fitness for chromosome in population]
    
    indices = np.argsort(scores) #this sorts each row from greatest to least and creates a list of indices
    
    return list(np.asarray(population)[indices])
```
